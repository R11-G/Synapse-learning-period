```python
t=[]
for i in range(0,891):
    name_eg=train_df.Name[i]
    title_search=re.search('(\w+)\.', name_eg)
    #print("result:", title_search)
    title = title_search.group(1)
    #print("title:", title)
    t.append(title)
train_df['Title']=t
train_df.head()
```
